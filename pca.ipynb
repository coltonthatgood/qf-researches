```python
import yfinance as yf
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
symbols = ['IBM', 'MSFT', 'META', 'INTC', 'AMD', 'NVDA', 'TSLA', 'GOOG']
data = yf.download(symbols, start="2020-01-01", end="2024-11-30")
print(data)
portfolio_returns = data['Close'].pct_change().dropna() #calculate portfolio returns based on pct change

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

pca = PCA(n_components=3)
principal_components = pca.fit_transform(scaled_data)

pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])

print("Explained Variance Ratio per Principal Component:")
print(pca.explained_variance_ratio_) 

print("\nCumulative Explained Variance:")
print(np.cumsum(pca.explained_variance_ratio_))

print("\nPrincipal Components (first 5 rows):")
print(pc_df.head())

print("\nComponents (eigenvectors):")
print(pca.components_)
```
